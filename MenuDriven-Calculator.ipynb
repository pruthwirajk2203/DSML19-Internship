```python
import CalculatorOperations as co
from IPython.display import clear_output
```

```python
another = 'Y'
print("Welcome to Calculator Operations!")
ans = 0
clear = 'N'
while another.upper() == 'Y':
    co.showMenu()
    print()
    calc_type = input("Please choose the Type of Calculator! : ")
    
    if calc_type == '1':
        co.basicMenu()
        print()
        opt = input("Please choose the operation! : ")
        
        if clear.upper() == 'Y':
            if opt in ['+','-','*','/','//','%','p']:
                num1, num2 = map(float,input("Enter two numbers").split())
            elif opt in ['s','c','sr']:
                num1 = float(input("Enter num1 : "))
            else:
                print("Please enter correct Operation!")
        elif clear.upper() == 'N':
            if ans == 0:
                if opt in ['+','-','*','/','//','%','p']:
                    num1, num2 = map(float,input("Enter two numbers").split())
                elif opt in ['s','c','sr']:
                    num1 = float(input("Enter num1 : "))
                else:
                    print("Please enter correct Operation!")
            else:
                if opt in ['+','-','*','/','//','%','p']:
                    num1 = ans
                    num2 = float(input("Enter num2 : "))
                elif opt in ['s','c','sr']:
                    num1 = ans
                else:
                    print("Please enter correct Operation!")
        else:
            clear = 'N'
        
        if opt == '+':
            ans = co.addition(num1, num2)
        elif opt == '-':
            ans = co.subtraction(num1, num2)
        elif opt == '*':
            ans = co.multiplication(num1, num2)
        elif opt == '/':
            ans = co.division(num1, num2)
        elif opt == '//':
            ans = co.floorDivision(num1, num2)
        elif opt == '%':
            ans = co.remainder(num1, num2)
        elif opt == 's':
            ans = co.square(num1)
        elif opt == 'c':
            ans = co.cube(num1)
        elif opt == 'p':
            ans = co.power(num1, num2)
        elif opt == 'sr':
            ans = co.squareRoot(num1)
        else:
            print("You have chosen wrong option!!")
        
    elif calc_type == '2':
        co.advanceMenu()
        print()
        opt = input("Please choose the operation! : ")
        
        if clear.upper() == 'Y':
            if opt in ['s','c','t','n','l','e','i','f','a','d','r']:
                num1 = float(input("Enter num1 : "))
            else:
                print("Please enter correct Operation!")
        elif clear.upper() == 'N':
            if opt in ['s','c','t','n','l','e','i','f','a','d','r']:
                num1 = ans
            else:
                print("Please enter correct Operation!")
        else:
            clear = 'N'
            
        if opt == 's':
            ans = co.sin_of(num1)
        elif opt == 'c':
            ans = co.cos_of(num1)
        elif opt == 't':
            ans = co.tan_of(num1)
        elif opt == 'n':
            ans = co.nat_log(num1)
        elif opt == 'l':
            ans = co.log_10(num1)
        elif opt == 'e':
            ans = co.exp_n(num1)
        elif opt == 'i':
            ans = co.inverse(num1)
        elif opt == 'f':
            ans = co.fact(num1)
        elif opt == 'a':
            ans = co.absolute(num1)
        elif opt == 'd':
            ans = co.deg(num1)
        elif opt == 'r':
            ans = co.rad(num1)
        else:
            print("You have chosen wrong option!!")
        
    else:
        print("You have chosen wrong option!!")
        
    print(ans)
    another = input("Do you want to perform another operation? \
                    (Enter Y to continue...)")
    clear = input("Do you want to clear current output? (Press Y to clear)")
    clear_output(wait=False)
```
