```python
import DictionaryOperations as do
from IPython.display import clear_output
```

```python
another = 'Y'
print("Welcome to Dictionary Operations!")
while another.upper() == 'Y':
    do.showMenu()
    print()
    opt = input("Please choose the option from list! : ")
    
    if opt == '1':
        myDict = do.createDictionary()
    elif opt == '2':
        print("Insert Item".center(20,'='))
        key = input("Enter the key for the item : ")
        item = do.generateStudentRecord()
        myDict = do.insertItem(myDict, key, item)
    elif opt == '3':
        print("Remove Item".center(20,'='))
        do.removeMenu()
        print()
        rem_opt = input("Please choose the option from list! : ")
        if rem_opt == 'a':
            key = input("Please enter the key to remove! : ")
            do.removeItem(myDict, key)
        elif rem_opt == 'b':
            do.removeLastInsertedItem(myDict)
        else:
            print("You have chosen wrong option!!")
    elif opt == '4':
        print("Search Item".center(20,'='))
        do.searchMenu()
        print()
        search_opt = input("Please choose the option from list! : ")
        if search_opt.upper() in 'SNME':
            searchValue = input("Please enter the search value! : ")
            searchBy = {'s':'Stud_Id', 'n':'Name', 'm':'Mobile_No', 'e':'Email_Id'}[search_opt]
            record = do.searchItem(myDict, searchBy, searchValue)
            print("Record : ",record)
        else:
            print("You have chosen wrong option!!")
    elif opt == '5':
        print("Sort Dictionary".center(20,'='))
        do.sortMenu()
        print()
        sort_opt = input("Please choose the option from list! : ")
        if sort_opt.upper() in 'SNME':
            if sort_opt.upper() == 'S':
                sortBy = 'Stud_Id'
            elif sort_opt.upper() == 'N':
                sortBy = 'Name'
            elif sort_opt.upper() == 'M':
                sortBy = 'Mobile_No'
            elif sort_opt.upper() == 'E':
                sortBy = 'Email_Id'
            else:
                print("You have chosen wrong option!!")
            val = input("Press 'a' for Ascending or 'd' for Descending : ")
            if val.upper() == 'A':
                sortType = False
            elif val.upper() == 'D':
                sortType = True
            else:
                print("You have chosen wrong option!!")
            myDict = do.sortDictionary(myDict, sortBy, sortType)
        else:
            print("You have chosen wrong option!!")
    else:
        print("You have chosen wrong option!!")
    
    print(myDict)
    another = input("Do you want to perform another operation? \
                    (Enter Y to continue...)")
    clear_output(wait=False)
```
